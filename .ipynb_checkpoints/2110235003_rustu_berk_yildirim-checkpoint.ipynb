# Kardiyovasküler Risk Oranı:
Kardiyovasküler risk, kişinin kalp ve damar hastalıklarına yakalanma olasılığını ifade eden bir kavramdır. Bu risk, genellikle bir dizi faktörün bir araya gelmesiyle belirlenir ve bireyin yaşam tarzı, genetik özellikleri ve sağlık durumu gibi birçok faktörü içerir.Genellikle yaş, cinsiyet, kan basıncı, kolesterol düzeyleri, sigara içme alışkanlığı, diyabet ve aile öyküsü gibi faktörleri içerir. Risk faktörleri yüksek olan bireylerde, önleyici önlemler ve yaşam tarzı değişiklikleri gibi tedbirler alınarak kardiyovasküler risk azaltılabilir. Bu, sağlıklı bir yaşam sürmek ve potansiyel sağlık sorunlarını önceden belirleyerek önlem almak için önemlidir.

```python
try:
    yas = int(input("Yaşınızı giriniz: "))
    total_kolesterol = float(input("Total kolesterol değerinizi giriniz: "))
    hdl_kolesterol = float(input("HDL kolesterol değerinizi giriniz: "))
    sistolik_bp = float(input("Sistolik BP değerinizi giriniz: "))
    sigara_kullanimi = input("Sigara kullanıyorsanız '1', kullanmıyorsanız '0' değerini giriniz: ")

    
    if yas < 40:
        yas_degeri = 0
    elif 40 <= yas < 50:
        yas_degeri = 1
    elif 50 <= yas < 60:
        yas_degeri = 2
    elif 60 <= yas < 70:
        yas_degeri = 3
    elif yas >= 70:
        yas_degeri = 4
    else:
        raise ValueError("Yaşınız 0 ile 100 arasında olmalıdır.Lütfen geçerli bir değer giriniz.")

    
    if total_kolesterol < 160:
        kolesterol_degeri = 0
    elif 160 <= total_kolesterol < 200:
        kolesterol_degeri = 1
    elif 200 <= total_kolesterol < 240:
        kolesterol_degeri = 2
    elif 240 <= total_kolesterol < 280:
        kolesterol_degeri = 3
    elif total_kolesterol >= 280:
        kolesterol_degeri = 4
    else:
        raise ValueError("Lütfen geçerli bir değer giriniz.")

    
    if hdl_kolesterol < 35:
        hdl_degeri = 4
    elif 35 <= hdl_kolesterol < 45:
        hdl_degeri = 3
    elif 45 <= hdl_kolesterol < 55:
        hdl_degeri = 2
    elif 55 <= hdl_kolesterol < 65:
        hdl_degeri = 1
    elif hdl_kolesterol >= 65:
        hdl_degeri = 0
    else:
        raise ValueError("Lütfen geçerli bir değer giriniz.")

    
    if sistolik_bp < 120:
        bp_degeri = 0
    elif 120 <= sistolik_bp < 130:
        bp_degeri = 1
    elif 130 <= sistolik_bp < 140:
        bp_degeri = 2
    elif 140 <= sistolik_bp < 160:
        bp_degeri = 3
    elif sistolik_bp >= 160:
        bp_degeri = 4
    else:
        raise ValueError("Lütfen geçerli bir değer giriniz.")

    
    if sigara_kullanimi == "1":
        sigara_degeri = 4
    elif sigara_kullanimi == "0":
        sigara_degeri = 0
    else:
        raise ValueError("Lütfen sigara kullanıyorsanız '1' kullanmıyorsanız '0' olarak giriniz.")

    
    risk_degeri = yas_degeri + kolesterol_degeri + hdl_degeri + bp_degeri + sigara_degeri
    risk = 1 - 0.988**(risk_degeri)

    
    print("Kardiyovasküler risk değeriniz:", "%",100*risk)

except ValueError():
    print("Lütfen geçerli bir değer giriniz.")
```

```python
def risk_Hesapla():
    try:
        yas = int(input("Yaşınızı giriniz: "))
        total_kolesterol = float(input("Total kolesterol değerinizi giriniz: "))
        hdl_kolesterol = float(input("HDL kolesterol değerinizi giriniz: "))
        sistolik_bp = float(input("Sistolik BP değerinizi giriniz: "))
        sigara_kullanimi = input("Sigara kullanıyorsanız '1', kullanmıyorsanız '0' değerini giriniz: ")

        if yas < 40:
            yas_degeri = 0
        elif 40 <= yas < 50:
            yas_degeri = 1
        elif 50 <= yas < 60:
            yas_degeri = 2
        elif 60 <= yas < 70:
            yas_degeri = 3
        elif yas >= 70:
            yas_degeri = 4
        else:
            raise ValueError("Yaşınız 0 ile 100 arasında olmalıdır. Lütfen geçerli bir değer giriniz.")

        if total_kolesterol < 160:
            kolesterol_degeri = 0
        elif 160 <= total_kolesterol < 200:
            kolesterol_degeri = 1
        elif 200 <= total_kolesterol < 240:
            kolesterol_degeri = 2
        elif 240 <= total_kolesterol < 280:
            kolesterol_degeri = 3
        elif total_kolesterol >= 280:
            kolesterol_degeri = 4
        else:
            raise ValueError("Lütfen geçerli bir değer giriniz.")

        if hdl_kolesterol < 35:
            hdl_degeri = 4
        elif 35 <= hdl_kolesterol < 45:
            hdl_degeri = 3
        elif 45 <= hdl_kolesterol < 55:
            hdl_degeri = 2
        elif 55 <= hdl_kolesterol < 65:
            hdl_degeri = 1
        elif hdl_kolesterol >= 65:
            hdl_degeri = 0
        else:
            raise ValueError("Lütfen geçerli bir değer giriniz.")

        if sistolik_bp < 120:
            bp_degeri = 0
        elif 120 <= sistolik_bp < 130:
            bp_degeri = 1
        elif 130 <= sistolik_bp < 140:
            bp_degeri = 2
        elif 140 <= sistolik_bp < 160:
            bp_degeri = 3
        elif sistolik_bp >= 160:
            bp_degeri = 4
        else:
            raise ValueError("Lütfen geçerli bir değer giriniz.")

        
        if sigara_kullanimi == "1":
            sigara_degeri = 4
        elif sigara_kullanimi == "0":
            sigara_degeri = 0
        else:
            raise ValueError("Lütfen sigara kullanıyorsanız '1' kullanmıyorsanız '0' olarak giriniz.")

        risk_degeri = yas_degeri + kolesterol_degeri + hdl_degeri + bp_degeri + sigara_degeri
        risk = 1 - 0.988**(risk_degeri)

        print("Kardiyovasküler risk değeriniz: ", "%",100*risk)

    except ValueError():
        print("Lütfen geçerli bir değer giriniz.")


risk_Hesapla()
```

```python
from tkinter import *
from tkinter import messagebox

class Pencere(Tk):
    def __init__(self):
        super().__init__()
        self.title("Kardiyovasküler Risk Hesaplama")
        self.geometry("400x400+250+250")
        self.araclar()
        self.menu()

    def araclar(self):
        self.etiket = Label(text="Risk Hesapla:")
        self.etiket.grid(row=0, column=0, columnspan=2, pady=5)

        self.label_yas = Label(text="Yaş:")
        self.label_yas.grid(row=1, column=0, padx=10, pady=5, sticky="e")
        self.yas = Entry()
        self.yas.grid(row=1, column=1, padx=10, pady=5)

        self.label_total_kolesterol = Label(text="Total Kolesterol:")
        self.label_total_kolesterol.grid(row=2, column=0, padx=10, pady=5, sticky="e")
        self.total_kolesterol = Entry()
        self.total_kolesterol.grid(row=2, column=1, padx=10, pady=5)

        self.label_hdl_kolesterol = Label(text="HDL Kolesterol:")
        self.label_hdl_kolesterol.grid(row=3, column=0, padx=10, pady=5, sticky="e")
        self.hdl_kolesterol = Entry()
        self.hdl_kolesterol.grid(row=3, column=1, padx=10, pady=5)

        self.label_sistolik_bp = Label(text="Sistolik BP:")
        self.label_sistolik_bp.grid(row=4, column=0, padx=10, pady=5, sticky="e")
        self.sistolik_bp = Entry()
        self.sistolik_bp.grid(row=4, column=1, padx=10, pady=5)

        self.label_sigara_kullanimi = Label(text="Sigara Kullanımı (1 veya 0):")
        self.label_sigara_kullanimi.grid(row=5, column=0, padx=10, pady=5, sticky="e")
        self.sigara_kullanimi = Entry()
        self.sigara_kullanimi.grid(row=5, column=1, padx=10, pady=5)

        self.hButon = Button(text="Hesapla", command=self.risk_Hesapla)
        self.hButon.grid(row=6, column=0, columnspan=2, pady=10)

        self.buton = Button(text="Kapat", command=self.destroy)
        self.buton.grid(row=7, column=0, columnspan=2, pady=5)

    def risk_Hesapla(self):
        try:
            yas = int(self.yas.get())
            total_kolesterol = int(self.total_kolesterol.get())
            hdl_kolesterol = int(self.hdl_kolesterol.get())
            sistolik_bp = int(self.sistolik_bp.get())
            sigara_kullanimi = self.sigara_kullanimi.get()

            if yas < 40:
                yas_degeri = 0
            elif 40 <= yas < 50:
                yas_degeri = 1
            elif 50 <= yas < 60:
                yas_degeri = 2
            elif 60 <= yas < 70:
                yas_degeri = 3
            elif yas >= 70:
                yas_degeri = 4
            else:
                raise ValueError("Yaşınız 0 ile 100 arasında olmalıdır. Lütfen geçerli bir değer giriniz.")

            if total_kolesterol < 160:
                kolesterol_degeri = 0
            elif 160 <= total_kolesterol < 200:
                kolesterol_degeri = 1
            elif 200 <= total_kolesterol < 240:
                kolesterol_degeri = 2
            elif 240 <= total_kolesterol < 280:
                kolesterol_degeri = 3
            elif total_kolesterol >= 280:
                kolesterol_degeri = 4
            else:
                raise ValueError("Lütfen geçerli bir değer giriniz.")

            if hdl_kolesterol < 35:
                hdl_degeri = 4
            elif 35 <= hdl_kolesterol < 45:
                hdl_degeri = 3
            elif 45 <= hdl_kolesterol < 55:
                hdl_degeri = 2
            elif 55 <= hdl_kolesterol < 65:
                hdl_degeri = 1
            elif hdl_kolesterol >= 65:
                hdl_degeri = 0
            else:
                raise ValueError("Lütfen geçerli bir değer giriniz.")

            if sistolik_bp < 120:
                bp_degeri = 0
            elif 120 <= sistolik_bp < 130:
                bp_degeri = 1
            elif 130 <= sistolik_bp < 140:
                bp_degeri = 2
            elif 140 <= sistolik_bp < 160:
                bp_degeri = 3
            elif sistolik_bp >= 160:
                bp_degeri = 4
            else:
                raise ValueError("Lütfen geçerli bir değer giriniz.")


            if sigara_kullanimi == "1":
                sigara_degeri = 4
            elif sigara_kullanimi == "0":
                sigara_degeri = 0
            else:
                raise ValueError("Lütfen sigara kullanıyorsanız '1' kullanmıyorsanız '0' olarak giriniz.")

            risk_degeri = yas_degeri + kolesterol_degeri + hdl_degeri + bp_degeri + sigara_degeri
            risk = 1 - 0.988**(risk_degeri)

            bilgiMetni = "Kardiyovasküler risk değeriniz: ", "%",100*risk
            messagebox.showinfo("Sonuç", bilgiMetni)

        except ValueError():
            messagebox.showerror("Hata", "Lütfen sadece sayısal değer giriniz.")

    def menu(self):
        self.menuBar = Menu(self)
        self.config(menu=self.menuBar)

        self.help = Menu(self.menuBar)
        self.help.add_command(label="kardiyovasküler risk nedir ?", command=self.yardim)
        self.help.add_command(label="Öğrenci Hakkında", command=self.ogrenciBilgi)
        self.menuBar.add_cascade(label="Hakkında", menu=self.help)

    def yardim(self):
        messagebox.showinfo("Yardım", """Kardiyovasküler risk, kişinin kalp ve damar hastalıklarına yakalanma olasılığını ifade eden bir kavramdır. Bu risk,yaş, cinsiyet, kan basıncı, kolesterol düzeyleri, sigara içme alışkanlığı, diyabet ve aile öyküsü gibi faktörlere göre değişir.""")

    def ogrenciBilgi(self):
        ogrenci_bilgi = "Öğrenci Bilgileri:\nAd: Rüştü Berk\nSoyad: Yıldırım\nÖğrenci No: 2110235003"
        messagebox.showinfo("Öğrenci Hakkında", ogrenci_bilgi)

pencere = Pencere()
pencere.mainloop()
```
